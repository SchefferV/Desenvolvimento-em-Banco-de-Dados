```python
---1

FUNCTION
---

CREATE TABLE customers ( id int, nome varchar(2));
insert into customers  ( id,nome) values(1,'A');
insert into customers ( id,nome) values(2,'B');


CREATE OR REPLACE FUNCTION totalCustomers
RETURN number IS
   total number(2) := 0;
BEGIN
   SELECT count(*) into total
   FROM customers;

   RETURN total;
END totalCustomers;
```

<h1> CURSOR </H1>


```python
--5
--5 É uma "cópia" da sua tabela que você pode manipular

CREATE TABLE CADASTRO ( CAD_id int, nome varchar(2));
insert into CADASTRO  ( CAD_id,nome) values(1,'A');
insert into CADASTRO ( CAD_id,nome) values(2,'B');

DECLARE
  CURSOR C_CADASTRO IS SELECT CAD_ID, nome from cadastro;
BEGIN
  FOR V_CADASTRO IN C_CADASTRO
  LOOP
    printf('cadastro:'|| V_CADASTRO.nome);
  END LOOP;
END;
```

```python
--6
--6  Faça um procedure que receba um número de ID e retorne se este ID está na tabela.



create or replace procedure buscarID(n1 IN number) IS
    CURSOR C_CADASTRO  IS SELECT CAD_ID, nome from cadastro where cad_id=n1 ;
    v_cad_idaux number;
BEGIN
    FOR V_CADASTRO IN C_CADASTRO
    LOOP
          printf('cadastro:'|| V_CADASTRO.nome);
    END LOOP;

    v_cad_idaux :=0 ;
    select cad_id into v_cad_idaux from cadastro where cad_id=n1 ;
    IF v_cad_idaux <= 0 or v_cad_idaux IS null THEN
        printf('Cadastro nao encontrado');
    END IF;
    EXCEPTION WHEN OTHERS THEN
        printf('Cadastro nao encontrado');
END;
```
